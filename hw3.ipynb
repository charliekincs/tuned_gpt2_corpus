```python
from transformers import Trainer, GPT2Tokenizer, TrainingArguments, DataCollatorForLanguageModeling, AutoModelForCausalLM
from datasets import load_dataset
import torch


dataset = load_dataset('text', data_files=['brothers_karamazov.txt', 'anna_karenina.txt', 'david_copperfield.txt', 'the_idiot.txt'])

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token
encoded_dataset = dataset.map(lambda examples: tokenizer(examples['text']))

training_args = TrainingArguments('train_output', save_strategy='no')
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
model = AutoModelForCausalLM.from_pretrained('gpt2')

trainer = Trainer(
    model,
    training_args,
    train_dataset=encoded_dataset['train'],
    data_collator=data_collator
)
trainer.train() 
trainer.save_model()
```
